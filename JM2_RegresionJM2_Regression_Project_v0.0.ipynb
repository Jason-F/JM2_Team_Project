LOADING DATA

```python

train=pd.read_csv('df-train_set.csv')
test=pd.read_csv('df-test_set.csv')
test_id=test['Index']
test=test.drop('Index',axis=1)
train.head()
```

```python
train=train[train['Commodities']=='APPLE GOLDEN DELICIOUS']
train=train.drop('Commodities',axis=1)
train=train.drop('Date',axis=1)
test=test.drop('Commodities',axis=1)
test=test.drop('Date',axis=1)
train=pd.get_dummies(train,drop_first=True)
test=pd.get_dummies(test,drop_first=True)
```

```python
train=pd.get_dummies(train,drop_first=True)
test=pd.get_dummies(test,drop_first=True)
```

```python
x=train.drop('avg_price_per_kg',axis=1)
y=train['avg_price_per_kg']
```

```python
test
```

```python
#x=train.drop('avg_price_per_kg_9.96',axis=1)
#y=train['avg_price_per_kg_9.96']
```

```python
# Split train dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, 
                                                    y, 
                                                    test_size=0.20,
                                                    random_state=1,
                                                    shuffle=False)
```

```python
decision_tree = DecisionTreeRegressor()
decision_tree.fit(X_train,y_train)
```

```python
train_DcT = decision_tree.predict(X_train)
test_DcT = decision_tree.predict(X_test)
```

```python
results_dict = {'Training MSE':{"Decision Tree": metrics.mean_squared_error(y_train, train_DcT)},
                'Test MSE':{"Decision Tree": metrics.mean_squared_error(y_test, test_DcT)},
                'Training R^2':{"Decision Tree": metrics.r2_score(y_train, train_DcT)},
                'Test R^2':{"Decision Tree": metrics.r2_score(y_test, test_DcT)}
                }
results=pd.DataFrame(results_dict)
```
